### 3. Librer√≠a matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.figure(figsize=(10, 6)) # Set figure size
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=2) # Plot with attribute

# Adding titles and labels
plt.title('Sine Wave Plot', fontsize=16) # Title with font size
```

```python
plt.xlabel('X-axis', fontsize=14) # X-axis label with font size
plt.ylabel('Y-axis', fontsize=14) # Y-axis label with font size

# Adding grid
plt.grid(True,linestyle='--', alpha=0.7) # Grid with style and transparency

# Adding legend
plt.legend(loc='upper right', fontsize=12) # Legend with location and font size

# Show the plot
plt.show()
```

### Code for a Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution

# Create a histogram
plt.figure(figsize=(10, 6))  # Set figure size
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)  # Histogram with attributes

# Adding titles and labels
plt.title('Histogram of Random Data', fontsize=16)  # Title with font size
plt.xlabel('Value', fontsize=14)  # X-axis label with font size
plt.ylabel('Frequency', fontsize=14)  # Y-axis label with font size

# Adding grid
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid on the y-axis with dashed lines

# Adding a vertical line for the mean
mean_value = np.mean(data)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')  # Mean line
plt.legend(loc='upper right', fontsize=12)  # Legend with location and font size

# Show the plot
plt.show()
```

### Code for a Pie Chart

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
sizes = [30, 25, 20, 15, 10]  # Corresponding sizes for each label
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet']  # Colors for each slice
explode = (0.1, 0, 0, 0, 0)  # Explode the first slice (Python)

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set figure size
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Adding a title
plt.title('Programming Language Popularity', fontsize=16)  # Title with font size

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  

# Show the plot
plt.show()
```

### Code for Distribution Chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)  # 1000 random numbers from a normal distribution

# Create a distribution plot
plt.figure(figsize=(10, 6))  # Set figure size
sns.histplot(data, bins=30, kde=True, color='skyblue', edgecolor='black', alpha=0.7)

# Adding titles and labels
plt.title('Distribution of Random Data', fontsize=16)  # Title with font size
plt.xlabel('Value', fontsize=14)  # X-axis label with font size
plt.ylabel('Frequency', fontsize=14)  # Y-axis label with font size

# Adding grid
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid on the y-axis with dashed lines
# Show the plot
plt.show()
```
